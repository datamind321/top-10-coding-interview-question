# Q1. Palindrome Numbers



- Palindrome String

```python
# palindrome with string

def is_palindrome(string):
  rev_str = string[::-1]
  if string == rev_str:
    print('Is Palindrome')
  else:
    print('Is Not Palindrome')

is_palindrome('sas')
```

```python
# 2nd method

def is_palindrome(string):
  n = len(string)   # get length of string
  for i in range(n):
    if string[i]==string[n-i-1]:
      return True
    else:
      return False

is_palindrome('nitin')
```

```python
# 3rd Method - using in-built function

def is_palindrome(string):
  rev = ''.join(reversed(string))
  if string == rev:
    print('Is Palindrome')
  else:
    print('Is Not palindrome')

is_palindrome('raj')
```

```python
# last 4th method using while loop

def is_palindrome(string):
  n = len(string)
  first = 0
  last = n-1
  while(first<last):
    if string[first]==string[last]:
      first += 1
      last -= 1
    else:
      return False
    return True

is_palindrome('nitin')
```

- Palindrome Number

```python
def is_palindrome_num(num):
  temp = num
  rev_num = 0
  while(temp>0):
    digit = temp%10
    rev_num = rev_num * 10 + digit
    temp = temp//10
  if num == rev_num:
    print('Is Palindrome Number')
  else:
    print('Is Not Palindrome Number')

is_palindrome_num(191)
```

# Q.3 Compress String

Input : aaabbccd

Output : a3b2c2d1

```python
# Using for loop
def compress_str(s):
  n = len(s)
  new_s = ''
  count = 1
  for i in range(n-1):
    if s[i] == s[i+1]:
      count += 1
    else:
      new_s = new_s + s[i] + str(count)
      count = 1

  new_s = new_s + s[n-1] + str(count)
  return new_s

s = 'aaabbccd'
compress_str(s)
```

# Q4. FizzBuzz


- if an number is divided by 3 then print 'fizz'
- if an number is divided by 5 then print 'buzz'
- if number divided by both 3 and 5 then print 'fizzbuzz'

```python
def fizzbuzz(n):
  for i in range(1,n+1):
    if i%3==0 and i%5 ==0:
      print('fizzbuzz')
    elif i%3==0:
      print('fizz')
    elif i%5==0:
      print('buzz')
    else:
      print(i)


fizzbuzz(25)
```

```python
# using dict

def fizzbuzz(n):
  d = {3:'fizz',5:'buzz'}
  for i in range(1,n+1):
    res = ''
    for k, v in d.items():
      if i % k == 0:
        res += v
    if not res:
      res = i
    print(res)

fizzbuzz(20)
```

# Q.5 Character Occurence

- Least repeating character in a string
- count of any particular element
- count of all elements

```python
# least repeating

def least_repeat(s):
  ch = {}
  for i in s:
    if i in ch:
      ch[i] = ch[i]+1
    else:
      ch[i]=1
  print(ch)
  result=min(ch,key=ch.get)
  print('least repeating character: ',result)

s='aaaabbbggghhi'
least_repeat(s)
```

```python
# count of any particular element

def count_char(s,search_ch):
  ch = {}
  for i in s:
    if i in ch:
      ch[i] = ch[i]+1
    else:
      ch[i]=1
  print(ch)
  print(ch[search_ch])

s='aaaabbbggghhi'
count_char(s,'h')
```

# Q.6 Prime Number

```python
def prime_num(n):
  flag = False
  if n>1:
    for i in range(2,n//2+1):
      if n%i==0:
        flag = True
        break
    if flag:
      return 'No Prime'
    else:
      return 'yes prime'


def prime_num2(n):
  if n>1:
    for i in range(2,n//2+1):
      if n%i==0:
        print('no')
        break
    else:
      print('yes')


# range of prime - particular interval range check prime or not

def check_prime(start,end):
  for n in range(start,end):
    if n>1:
      for i in range(2,n//2+1):
        if n%i==0:
          break
      else:
        print(n)

prime_num2(53)
check_prime(1,20)
```

# Q.8 Find Second Largest Number

```python
# using for loop

def second_large(n):
  if n[0]>n[1]:
    first = n[0]
    second = n[1]
  else:
    first = n[1]
    second=n[0]
  for i in range(2,len(n)):
    if n[i]>first:
      second=first
      first = n[i]
    elif n[i]>second and first != n[i]:
      second=n[i]
  return second

# using reverse sort

def sec_large_rev(l):
  l.sort(reverse=True)
  return l[1]

# using set and max

num = [10,20,30,40,50]

new_set = set(num)
new_set.remove(max(new_set))
print('Using Set and Max: ',max(new_set))

print('Using List: ',second_large(num))
print('Using Reverse Sort: ',sec_large_rev(num))
```

# Q.9 Armstrong Number

```python
def armstrong_number(n):
  sum=0
  temp=n
  while temp>0:
    digit = temp%10
    sum = sum + (digit**3)
    temp = temp//10
  if n == sum:
    print('y')
  else:
    print('n')


def get_armstrong_number(start,end):
  for n in range(start,end):
    sum=0
    temp=n
    while temp>0:
      digit = temp%10
      sum = sum + (digit**3)
      temp = temp//10
    if n == sum:
      print(n)


def sum_of_digits(n):
  sum=0
  temp=n
  if n>9:
    while temp>0:
      digit = temp%10
      sum = sum + digit
      temp = temp//10
  else:
    sum=n
  return sum


armstrong_number(153)
print('--------- range of prime numbers ----------')
get_armstrong_number(0,500)
print('sum of digits: ',sum_of_digits(12144))
```

# Q.10 Factorial Number

```python
def factorial(n):
  fact = 1
  if n <= 0:
    print('Invalid')
  else:
    for i in range(1,n+1):
      fact = fact*i
    return fact

factorial(7)
```

- sum of factorial

eg. 145! = 1! + 4! + 5! = 1+24+120 = 145

```python
def sum_of_fact(n):
  temp=n
  sum=0
  while temp>0:
    digit = temp%10
    temp = temp//10
    fact = factorial(digit)
    sum = sum+fact
  if sum==n:
    print('y')
  else:
    print('n')

sum_of_fact(145)
```

```python
# particular interval get sum_of_factorial
def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact = fact*i
    return fact


def list_of_sum_fact(start,end):
  for n in range(start,end):
    temp=n
    sum=0
    while temp>0:
      digit = temp%10
      temp = temp//10
      fact = factorial(digit)
      sum = sum + fact
    if sum == n:
      print(n)

list_of_sum_fact(1,2000)
```
